### <font color="green"> ***Exercise: Simulate a random DNA sequence*** </font>

* with a equal base frequencies
* with a SARS-CoV-2 genome, which contains 29,903 base pairs (bps)and a base composition of 29.94% of A, 18.37% of C, 19.61% of G, and 32.08% of T




```python
import random
bases = list("atcg")
genome_size = 1000
dna = random.choices(bases, k=genome_size) # sample with replacement
print("".join(dna))

ct_g = dna.count("g")
ct_c = dna.count("c")
print((ct_g + ct_c)*100/genome_size)
```

```python
dna_biased = random.choices(list("acgt"), k=genome_size, weights=[0.299, 0.184, 0.196, 0.321]) # sample with biased base frequencies
print("".join(dna_biased))

ct_g = dna_biased.count("g")
ct_c = dna_biased.count("c")
print((ct_g + ct_c)*100/genome_size)
```

### <font color="green"> ***Exercise: DNA replication*** </font>

Write a function that simulates the DNA replication.
*   Input: a single strand of DNA (in the 5 to 3 prime direction)
*   Output: a daughter DNA string (also in the 5 to 3 prime direction)





```python
seq="aaattaccgggaaattgc" # the actual DNA is double-stranded; rev-com (3 to 5 primer direction) is not presented
daughter_strand = "tttaatgccctttaacg" # base pairing rule: G-C, A-T (3 to 5)
expected_strand = "gcaatttcccgtaattt" # reverse it!!!
expected_rna_transcript = "gcaauuucccguaauuu" # base pairing rule: G/C; A/U
```

### solution 1: use if_else

```python
# make a function
def dna_replication_loop(seq: str) -> str:
    seq = seq.upper()
    com_strand = list() #  initialize an empty list (alternatively, initialize a null string, and then concatenate with "+")
    for base in list(seq):
        if base == 'A':
            com_strand.append("T") # append the list
        elif base == 'T':
            com_strand.append("A")
        elif base == 'C':
            com_strand.append("G")
        elif base == 'G':
            com_strand.append("C")
        else:
            com_strand.append("N")
    com_strand.reverse() # reverse the list
    rev_com = ''.join(com_strand)
    print(seq, "\n", rev_com, "\n")
    return(rev_com)

dna_replication_loop("ATTTTAcccatgg")
```

### Solution 2. Use dictionary

```python
# use a dictionary (to avoid ifelse testing)
# and list comprehension
def dna_replication_dict(seq: str) -> str:
    seq = seq.upper()
    comp_bases = {
        "A": "T",
        "T": "A",
        "C": "G",
        "G": "C"
        }
    # com_strand = [comp_bases[base] for base in list(seq)] # one-liner for if_else conditional testing
    # to be safe (non-standard bases)
    com_strand = list()
    for base in list(seq):
        if base in comp_bases: # test if the key exists or not
            com_strand.append(comp_bases[base])
        else: # key not exist; otherwise throw an exception (error)
            com_strand.append("N") # N for undefined base (non-canonical bases; R=>A/G; Y=>C/T; ambiguous bases)

    com_strand.reverse() # reverse the list
    rev_com = ''.join(com_strand)
    print(seq, "\n", rev_com, "\n")
    return(rev_com)

dna_replication_dict("ATTTTAcccrrxatgg")
```

```python
def dna_replication_bio(seq: str) -> str:
    from Bio.Seq import Seq # import Bio module
    dna = Seq(seq) # turn the string to a Seq object
    revcom = dna.reverse_complement() # rev comp; takes care of ambiguous bases: R (A/G) <=> Y (C/Y)
    return("".join(str(revcom))) # turn Seq object back to string and join

dna_replication_bio("ATTTTAcccrrnwatgg")
```
