### <font color="green"> ***Exercise: Simulate DNA mtuations with transition/transversion bias*** </font>

* Given a mutation rate of mu=10e-3 per site per generation, the **expected** number of mutations n=mu * genome_size
* For **actual** number of mutations, k, is distributed as a Poisson(n)
* For each mutation, we pick a genome position randomly (uniform distribution) and ...
    ** replace its base with a transition (Ti) or transversion (Tv), with the Ti 5 times as frequent as Tv





```python
def GC_percent(dna:str) -> float:
    dna = dna.upper()
    dna_length = len(dna)
    ct_g = dna.count("G")
    ct_c = dna.count("C")
    return((ct_g + ct_c) * 100/dna_length)
```

```python
import random
import textwrap
# define a function with default arguments (and all input & output data types)
def rand_dna(genome_size:int = 1000, base_freq:list = [1,1,1,1]) -> str: 
    bases = list("atcg")
    dna = random.choices(bases, k = genome_size, weights = base_freq) # sample with replacement
    return("".join(dna))

# Test 1: equal base freqs:
seq_out = rand_dna()
print(textwrap.fill(seq_out))
GC_percent(seq_out)

# Test 2. an AT-rich genome (like ticks and Lyme bacteria)
seq_out = rand_dna(base_freq = [3,3,1,1])
print(textwrap.fill(seq_out))
GC_percent(seq_out)


```

```python
# default mutation rate 10^-3
import numpy as np
rng = np.random.default_rng() # initialize a random number genenerator

# use the broken stick model to simulate a biased outcome:
def mutate_dna(seq:str, mut_rate:float = 1e-3, ti_tv:float = 5.0) -> str:
  print("random seq:", textwrap.fill(seq))
  seq = seq.upper()
  seq_list = list(seq)
  bases = list("ATCG")
  seq_length = len(seq)
  mut_exp = seq_length * mut_rate
  num_obs = rng.poisson(mut_exp) # return an interger from a Poisson distribution with the mean defined above

  transitions = {"A": "G", "G": "A", "C": "T", "T": "C"}  

  for i in range(num_obs): # for each mutation
    pos = rng.integers(low=0, high=seq_length-1) # a uniform random position, zero-indexed
    print("mutation number:", i, "at position:", pos + 1, end="\t") # add 1 since python is 0-based indexing
    original_base = seq_list[pos]
    possible_bases = [b for b in bases if b != original_base] # list comprehension, a one-liner shorthand for if-else control loop
    broken_stick = rng.uniform() # returns a random number between 0 and 1
    new_base = ''
    if broken_stick < 1/(ti_tv + 1): # a transversion, with prob 1/6
      new_base = rng.choice([b for b in possible_bases if b != transitions[original_base]])
    else: # transition, with prob 5/6
      new_base = transitions[original_base]
    print("original base:", original_base, "new base:", new_base)
    seq_list[pos] = new_base
  return "".join(seq_list)

# Test with a random seq
seq_out = rand_dna()
print ("Mutate_DNA:", textwrap.fill(mutate_dna(seq_out, 1e-2)))
```

### <font color="green"> ***Exercise: DNA translation, in all six frames*** </font>

Write a function that simulates the DNA translation.
*   Input: a single strand of DNA (in the 5 to 3 prime direction)
*   Output: translated protein sequences in all six possible frames





```python
# reverse complement
def revcom(seq: str) -> str:
    seq = seq.upper()
    comp_bases = {
        "A": "T",
        "T": "A",
        "C": "G",
        "G": "C"
        }
    com_strand = list()
    for base in list(seq):
        if base in comp_bases: # test if the key exists or not
            com_strand.append(comp_bases[base])
        else: # key not exist; otherwise throw an exception (error)
            com_strand.append("N") # N for undefined base (non-canonical bases; R=>A/G; Y=>C/T; ambiguous bases)

    com_strand.reverse() # reverse the list
    rev_com = ''.join(com_strand)
#    print(seq, "\n", rev_com, "\n")
    return(rev_com)

revcom("ATTTTAcccrrxatgg")
```

```python
def translate_a_dna(seq:str, frame:int) -> str:
    # Represent the genetic code as a dictionary
    codons = ["TTT","TCT","TAT","TGT","TTC","TCC","TAC","TGC","TTA","TCA","TAA","TGA","TTG", "TCG","TAG","TGG","CTT","CCT","CAT","CGT","CTC","CCC","CAC","CGC","CTA","CCA", "CAA","CGA","CTG","CCG","CAG","CGG","ATT","ACT","AAT","AGT","ATC","ACC","AAC", "AGC","ATA","ACA","AAA","AGA","ATG","ACG","AAG","AGG","GTT","GCT","GAT","GGT", "GTC","GCC","GAC","GGC","GTA","GCA","GAA","GGA","GTG","GCG","GAG","GGG"]
    aas = ["F","S","Y","C","F","S","Y","C","L","S","*","*","L","S","*","W","L","P","H","R", "L","P","H","R","L","P","Q","R","L","P","Q","R","I","T","N","S","I","T","N","S", "I","T","K","R","M","T","K","R","V","A","D","G","V","A","D","G","V","A","E","G","V","A","E","G"]
    genetic_code = dict(zip(codons, aas))
    seq = seq.upper() # make sure all caps

    pep = []
    for i in range(abs(frame)-1, len(seq)-2, 3): # note use the absolute value of frame; also to use zero-based position counting
        codon = seq[i:i+3]
        if codon in genetic_code:
            aa = genetic_code[codon] 
        else:
            aa = 'X' # undefined amino acid (e.g., codon contains non-ATCG letter like "N", "Y", "R")
        pep.append(aa)

    return("".join(pep))
```

```python
# test with a random DNA of 100 bases
seq_out = rand_dna(genome_size = 100)
print(seq_out)
print("frame +1", translate_a_dna(seq_out, 1))
print("frame +2", translate_a_dna(seq_out, 2))
print("frame +3", translate_a_dna(seq_out, 3))
print("frame -1", translate_a_dna(revcom(seq_out), 1))
print("frame -2", translate_a_dna(revcom(seq_out), 2))
print("frame -3", translate_a_dna(revcom(seq_out), 3))

from Bio.Seq import Seq # import Bio module
dna = Seq(seq_out) # turn the string to a Seq object
revcom = dna.reverse_complement() # rev comp; takes care of ambiguous bases: R (A/G) <=> Y (C/Y)
print(dna.translate())


```
