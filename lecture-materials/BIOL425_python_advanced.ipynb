```python
# here's our first function
def GCcontent(dna: str) -> float:
  # our function is called GCcontent and
  # accepts a single argument called dna (data type: string)
  # The output (GC%) data type is a floating point number
  # assume that the input is a DNA sequence encoded
  # in a string, and make sure it's all uppercase:
  dna = dna.upper()
  # count the occurrences of each nucleotide
  numG = dna.count("G")
  numC = dna.count("C")
  numA = dna.count("A")
  numT = dna.count("T")
  # finally, calculate (G + C) / (A + T + G + C)
  return (numG + numC) / (numG + numC + numT + numA)
```

```python
# print a dictionary
# no return values
def print_dictionary(mydic):
  for k, v in mydic.items():
    print("key: ", k, " value: ", str(v))

# return a list with results
# declare default arguments: if no input is provided,
# assume start = 1, end = 10 (as defaults)
def squared(start = 1, end = 10):
  # create empty list to catch result of each cycle
  results = []
  for i in range(start, end):
    r = i ** 2
    # append current value to result list
    results.append(r)
  return results
```

```python
# Q3. sort three nubmers
def foo3(x = 2, y = 0, z = 9):
  if x > y:
    tmp = y
    y = x
    x = tmp
  if y > z:
    tmp = z
    z = y
    y = tmp
  if x > y:
    tmp = y
    y = x
    x = tmp
  return [x, y, z]
```

```python
# Q5: find all prime factors (or itself if it is a primer number)
def foo5(x = 1729):
  d = 2
  myfactors = []
  while x > 1:
    if x % d == 0:
      myfactors.append(d)
      x = x / d
    else:
      d = d + 1
  return myfactors
```

```python
# Q6:  factorial, recursive & multiple down
# foo6 is a recursive function, meaning that the function calls itself (until a termination condition)
def foo6(x = 25):
  if x == 1:
    return 1
  return x * foo6(x - 1)
```

```python
# Q7: find all prime numbers up to a number
def foo7(x = 100):
  myp = [2] # keep a list of prime numbers
  for i in range(3, x + 1):
    success = False
    for j in myp:
      # print(i, j)
      if i % j == 0: # if divisible by a prime, it's not a prime. Skip
        success = True
        break
    if success == False: # find a prime
        myp.append(i)
  return myp
```

```python
print(foo7())
```

```python
# generate a random dna sequence
#  with default of 10 bases
def rand_dna(dan_len: int) -> str:
  import numpy as np
  rand_seq = []
  rng = np.random.default_rng()
  for i in range(dan_len):
    rand_seq.append(rng.choice(['A','C','G','T']))
  return(''.join(rand_seq))
```

### <font color="green"> ***Exercise: DNA mutations*** </font>
Mutations are randomly introduced during DNA replication due to misincorporation of bases by DNA polymerase (even with its proof-reading abilities).
* The mutation positions can be simulated with a random number drawn from a uniform distribution of DNA positions
* The mis-incorporated base can be simulated with a random choice of bases. Mutated base is not uniform, with **transitions** (A<=>G, and C<=>T) more often than **transversion** (purines and pyrimidines)
* Mutation rate: ~$10^{-9}$ per base per replication
* Mutations provide the raw sources of genetic variability and natural selection



```python
import random
import textwrap
def mutate_dna(seq: str, num_mut: int) -> str: # define the inputs and output (with data types)
  print("random seq:", textwrap.fill(seq))
  seq_list = list(seq)
  bases = ["A", "T", "C", "G"]
  for i in range(num_mut): # for each mutation
    pos = random.randint(0, len(seq_list)-1) # a uniform random position
    print("mutation number:", i+1, "at position:", pos + 1, end="\t") # add 1 since python is 0-based indexing
    original_base = seq_list[pos]
    possible_bases = [b for b in bases if b != original_base] # list comprehension, a one-liner shorthand for if-else control loop
    new_base = random.choice(possible_bases) # a uniform random base
    print("original base:", original_base, "new base:", new_base)
    seq_list[pos] = new_base
  return "".join(seq_list)
print ("Mutate_DNA:", textwrap.fill(mutate_dna(rand_dna(100),5)))
```

### <font color="green">**Exercise: six-frame translation**</font>
* Not all DNA sequences code for a protein. In fact, < 5% DNA in the human genomes code for protein (most DNAs in eukaryotes are non-coding, including introns, repetitive elements, and viral remanents)
* **Codon**:  On a protein-coding DNA sequence, a combination of 3 bases makes a codon. There are a total of 64 possible codons. Strictly speaking, codons refer to triplets on tRNA (transfer RNA), so it should be RNA based. But here we will use DNA-based notation to refer to codons.
* **Genetic code**: The 64 possible codons encode a total of 20 amino acids and (translation) stop codons. We will implement genetic code as a dictionary
* Reading frames:
  * a protein-coding sequence can be translated into three possible peptide sequences depending on the starting position. There are the three positive reading frames: +1, +2, and +3.
  * The same DNA sequence has an un-written rev-complement strand, which may be the actual coding strand. The rev-com strand can also be translated into three possible peptide sequences, called -1, -2, and -3 reading frames
